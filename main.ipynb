```python
import cv2
from google.colab.patches import cv2_imshow

import numpy as np
```

```python
# Load the Haar Cascade XML file for face detection
# face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
# Specify the path to your custom Haar Cascade XML file
custom_cascade_path = "/haar_cascade_face_detection.xml"

# Load your custom Haar Cascade XML file for face detection
face_cascade = cv2.CascadeClassifier(custom_cascade_path)
```

```python
# Load your grayscale image
image_path = '/dataset/dataset1.jpg'
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Get the resolution of the image
height, width = gray.shape[:2]
```

```python
# Display the result of grayscale image processing
cv2_imshow(gray)

# Add the resolution information
text = f"Resolution: {width}x{height}"
print(text)
```

```python
# Perform face detection
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

# Draw rectangles around the faces and add text labels
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 3)
    cv2.putText(image, 'Face', (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)
```

```python
# Display the result
cv2_imshow(image)
```
